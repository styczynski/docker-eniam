```python
# !pip install opencv-python
# !pip install camelot-py
# !pip install wand
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from eniam import Eniam, rule, root_rule, valence_rule

dom_result = Eniam(['KOT', 'MYSZ'], {
    rule(lemma='kot',pos='subst',case='nom'):     'KOT',
    rule(lemma='goniƒá',pos='fin',person='ter'):   '(ip\\KOT)/MYSZ',
    rule(lemma='mysz',pos='subst',case='acc'):    'MYSZ',
    root_rule():                                  's\\?(ip)',
}, ['KOT', 'MYSZ', 'ZDARZENIE'], {
    valence_rule('kot', 'noun'): 'KOT',
    valence_rule('mysz', 'noun'): 'MYSZ',
    valence_rule('goniƒá', 'verb'): 'ZDARZENIE',
}).dom("Kot goni mysz.")

dom_result.show()
```
